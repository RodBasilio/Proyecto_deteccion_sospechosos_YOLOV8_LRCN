```python
import cv2
import numpy as np
import os

# Directorio de las imágenes originales y las etiquetas
imagenes_dir = "prueba 2/data/images/train"
labels_dir = "prueba 2/data/labels/train"

# Función para cargar imágenes y etiquetas
def cargar_imagenes_y_etiquetas(imagenes_dir, labels_dir):
    imagenes = []
    etiquetas = []

    # Obtener la lista de nombres de archivos en ambas carpetas
    nombres_imagenes = os.listdir(imagenes_dir)
    nombres_etiquetas = os.listdir(labels_dir)

    # Iterar sobre los nombres de archivos de las imágenes
    for nombre_imagen in nombres_imagenes:
        # Comprobar si el archivo de etiqueta correspondiente existe
        if nombre_imagen in nombres_etiquetas:
            # Cargar la imagen
            imagen = cv2.imread(os.path.join(imagenes_dir, nombre_imagen))

            # Leer las etiquetas desde el archivo de etiquetas
            with open(os.path.join(labels_dir, nombre_imagen), "r") as archivo_etiquetas:
                etiquetas_texto = archivo_etiquetas.read()
                etiquetas = [float(valor) for valor in etiquetas_texto.split()]

            imagenes.append(imagen)
            etiquetas.append(etiquetas)

    return imagenes, etiquetas

# Función para dibujar puntos en la imagen
def dibujar_puntos(imagen, etiquetas):
    imagen_con_puntos = imagen.copy()

    for i in range(0, len(etiquetas), 2):
        x = int(etiquetas[i] * imagen.shape[1])
        y = int(etiquetas[i + 1] * imagen.shape[0])
        cv2.circle(imagen_con_puntos, (x, y), 5, (0, 255, 0), -1)  # Dibuja un círculo en el punto

    return imagen_con_puntos

# Cargar imágenes y etiquetas
imagenes, etiquetas = cargar_imagenes_y_etiquetas(imagenes_dir, labels_dir)

# Visualizar imágenes con puntos de etiqueta
for i, (imagen, etiquetas) in enumerate(zip(imagenes, etiquetas)):
    imagen_con_puntos = dibujar_puntos(imagen, etiquetas)
    cv2.imshow(f"Imagen {i}", imagen_con_puntos)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
